### 문제 (3)
1. PM10 농도 컬럼과 CO2 농도 컬럼의 최소-최대 정규화를 시행 , 각 컬럼의 표준편차를 구해라 
2. PM10 농도 컬럼의 표준편차가 a이고 CO2 농도 컬럼의 표준편차가 b 일때 a - b 값을 구하시오 
(단, 최대최소 정규화는 X = (xi -xmin)/(xmax-xmin))와 같이 연산하고  연산결과는 반올림 해서 소수점 4자리 수 까지 출력한다.

```python
import pandas as pd
import numpy as np # np.std() 표준편차 확인 가능 
data = pd.read_csv('https://raw.githubusercontent.com/JEunJin/Bigdata_python/refs/heads/master/bigdata_csvfile/Environmental%20data.csv')
#data.head()
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
data['Co2 농도'] = mm.fit_transform(data[['Co2 농도']])
data['PM10 농도'] = mm.fit_transform(data[['PM10 농도']])
print(np.std(data['Co2 농도']))  # ddof =0 인 표준편차 계산  차이가 있으니 문제에 유의할 것
print(np.std(data['PM10 농도'])) # ddof =0 인 표준편차 계산
print(round(data['Co2 농도'].std(ddof=0),4),round(data['PM10 농도'].std(ddof = 0),4)) # 0.3162 , 0.2763  # ddof = 0 (default)인 표준편차 계산 
```

## 작업형 2유형 
- 지하철 이용객에 대한 데이터셋 
1. 전체 데이터의 상위 5000행을 train 데이터로 사용하고, 나머지를 test 데이터로 사용할 수 있도록 데이터 슬라이싱 작업을 한 후에 데이터를 분석하시오 
2. 제공된 train 데이터 활용하여 '이용 횟수'를 예측하는 모델을 개발하고, 개발한 모델에 기반하여 평가용 test 데이터에 적용하여 얻은 '이용 횟수' 예측 결과를 csv 파일로 생성하는 코드를 제출하시오 
3. 제출 결과는 test 데이터 개수인 1000행이 될 수 있도록 한다.
4. 제출 데이터 컬럼은 이용 횟수만 존재해야 하고 , 평가지표는 MAE로 한다.
5. 예측 결과 파일명은 8th_test_type2.csv

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/JEunJin/Bigdata_python/refs/heads/master/bigdata_csvfile/subway_usage_data_with_payment.csv')
#data.head()
train = data.iloc[0:5000,:]
test = data.iloc[5000:,:] # 1000개
#display(test.head()) 
# 이용횟수 연속형 변수 >  회귀예측 , MAE 평가지표 
# 성별 , 나이 , 직업 , 호선 , 지불방법 >> 연속형 변수로 변환 필요 
from sklearn.preprocessing import LabelEncoder
cols = ['성별','나이','직업','호선','지불방법']
le = LabelEncoder()
for c in cols:
    train[c] = le.fit_transform(train[c])
    test[c] = le.fit_transform(test[c])
display(test.head()) 

from sklearn.model_selection import train_test_split
X = train.drop(columns=['이용 횟수'])
y = train['이용 횟수']
X_train , X_val ,y_train , y_val = train_test_split(X,y,test_size=0.2)

from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(X_train,y_train)
y_pred=rfr.predict(X_val)

from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_pred, y_val))

# 예측 해보기 
X_test = test.drop(columns=['이용 횟수'])
y_pred2 = rfr.predict(X_test)

pd.DataFrame({'이용 횟수':y_pred2}).to_csv('8th_test_type2.csv',index=False)
result = pd.read_csv('8th_test_type2.csv')
print(result)
```

### 작업형 3유형 (1)
- 고객의 이탈여부 Churn을 예측하는 문제제
1. 로지스틱 회귀분석 실시, 유의하지 않은 변수의 개수를 구해라 
2. 유의미한 변수들만 독립변수로 하여 로지스틱 회귀분석 재실시 , 회귀계수의 평균 구하시오 
3. 적용한 회귀식에서 tenure 변수가 5증가할 때 오즈비를 구해라 (소수점 3자리 )


```python
import pandas as pd 
import numpy as np
data = pd.read_csv('https://raw.githubusercontent.com/JEunJin/Bigdata_python/refs/heads/master/bigdata_csvfile/Telco_customer_churn_data.csv')
#data.info() # 전부 연속형, 로지스틱 분석 가능 

import statsmodels.formula.api as sm

result1 = sm.logit('Churn ~ gender + SeniorCitizen + Dependents + tenure + PhoneService + OnlineSecurity + MonthlyCharges',data=data).fit().summary()
#print(result1) 
# 1. 유의하지 않은 변수 개수 : gender 1개 
print(1)
# 2. 유의미한 변수들로만 로지스틱 회귀분석 재실시 
result2 = sm.logit('Churn ~ SeniorCitizen + Dependents + tenure + PhoneService + OnlineSecurity + MonthlyCharges',data=data).fit()
print(round(result2.params.mean(),3))  # -0.234
# 3. 적용한 회귀식에서 tenure 가 5 증가할 때 오즈비 
odd_5 = np.exp(result2.params['tenure']*5)
print(round(odd_5,3))

 
```

### 작업형 3유형 (2)
- 지능지수(PIQ)를 예측
1. 다중 선형 회귀분석 적용,가장 유의미한 변수의 회귀계수를 구하시오 
2. 1번 모델의 결정계수의 값을 구하시오
3. 키 70 , 몸무게 150 , 뇌크기 90일 때 PIQ 값을 구하시오 
(소수점 3자리 까지 반올림)


```python
import pandas as pd 
data = pd.read_csv('https://raw.githubusercontent.com/JEunJin/Bigdata_python/refs/heads/master/bigdata_csvfile/PIQ_Brain_Height_Weight.csv')
#data.info() 전부 연속형 변수 
import statsmodels.api as sm 
X = data.drop(columns='PIQ')
y = data['PIQ']
X = sm.add_constant(X)
result1 =sm.OLS(y,X).fit()
print(result1.summary()) # 가장 유의미한 회귀변수(Brain)
# 문제 1  회귀계수
print(round(result1.params['Brain'],3))
# 문제 2 결정계수 
print(0.295)
# 문제 3 예측
y_pred = result1.predict([1,90,70,150])
print(round(y_pred[0],3))
```

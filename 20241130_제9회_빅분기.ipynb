### 문제 (1)
1. '대출 A' 와 '대출 B'칼럼을 합하여 '총대출액' 컬럼을 만든다 
2. 성별간 총 대출액 차이의 절대값이 가장 큰 지역 코드를 구하시오 


```python
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/JEunJin/BigData_python/refs/heads/master/bigdata_csvfile/loan_data.csv')
data.head()
data['총대출액'] = data['대출A']+data['대출B']

#**
result = data.groupby(['지역코드','성별'])['총대출액'].sum()
print(result)
# 성별 << 같은 지역코드 성별간 차이를 알기 위헤 멀티 인덱스 컬럼을 unstack()화 
result = result.unstack()
print(result)

result['지역간 차이'] = abs(result[0]-result[1])
display(result.sort_values('지역간 차이',ascending=False))
print(1000009974)
```

### 문제 (2)
1. 각 연도별로 각 범죄의 검거건수를 발생건수로 나눈 검거율을 계산
2. 연도별로 검거율이 가장 높은 범죄형태의 검거건수의 총합을 계산 


```python
import pandas as pd 
data = pd.read_csv('https://raw.githubusercontent.com/JEunJin/BigData_python/refs/heads/master/bigdata_csvfile/crime_data_2014_2020.csv')

발생건수 = data.iloc[::2].set_index('년도').drop(columns='구분')
검거건수 = data.iloc[1::2].set_index('년도').drop(columns='구분')

검거율 = 검거건수 / 발생건수
display(검거율)

total = 0
# 연도별로 검거율이 가장 높은 범죄형태의 검거건수 총합 계산 
for year in 검거율.index:
    highest = 검거율.loc[year].idxmax()# 검거율이 가장 높은 범죄명을 highest 에 대입
    total += 검거건수.loc[year , highest]

print(total)
```

### 문제 (3)
1. 결측값이 존재하는 컬럼 2개 중 '연봉'은 해당 컬럼의 평균으로 대체 , '근속년수'는 '부서명'과 '근무평가등급'이 같은 직원의 '근속년수' 평균으로 대체 
2. 연봉을 근속년수로 나눈 컬럼과 , 연봉을 고객만족점수으로 나눈 컬럼을 파생변수로 생성 , 각 파생변수를 기준으로 데이터를 정렬 
3. 두개의 파생변수를 기준으로 내림차순 정렬한 가각의 데이터에서 세번째로 높은 데아터의 '근속년수'를 a , 두번째로 높은데아터의 '교육참가시간수'를 b 라고 할때 a+b를 구하시오ㅓ 
(단, 출력값을 정수 형태로 제출)

```python
import pandas as pd 
data = pd.read_csv('https://raw.githubusercontent.com/JEunJin/BigData_python/refs/heads/master/bigdata_csvfile/employee_data_with_grades.csv')
#data.head()
data['연봉'].fillna(data['연봉'].mean(),inplace=True)
# '부서명'과 '근무평가등급'에 대한 '근속년수'의 평균으로 결측값 채우기
data['근속년수'] = data.groupby(['부서명','근무평가등급'])['근속년수'].transform(lambda x : x.fillna(x.mean())) # 각 그룹 내 평균값으로 결측값을 대체 

data.head()
data['연봉/근속년수'] = data['연봉']/data['근속년수']
data['연봉/고객만족점수'] = data['연봉']/data['고객만족점수']

data1 = data.sort_values('연봉/근속년수',ascending=False)
data2 = data.sort_values('연봉/고객만족점수',ascending=False)
display(data1.head()) # EMP0002 > 근속년수 : 1.0
display(data2.head()) # EMP0350 > 교육참가시간수 : 80
a = data1['근속년수'].iloc[2]
b = data2['교육참가시간수'].iloc[1]
print(int(a+b))
```

### 작업형 2유형
1. 주어진 데이터 상위 700행을 train 데이터로 / 나머지를 test로 사용 
2. 제공된 데이터를 이용하여 '운동 경험 수준(Experience_Level)'을 예측하는 모델 개발 , 얘측 결과를 csv 파일로 셍성하여 제출
3. 예측결과는 f1-score macro 평가지표에 따라 평가하고, 평가용 데이터(test)개수와 예측결과 데이터의 개수는 273개로 일치해야 함 
4. 예측 결과 파일명은 9th_test_type2.csv로 하고 컬럼 명은 pred 하나만 존재하도록 한다 

```python
import pandas as pd 

data = pd.read_csv('https://raw.githubusercontent.com/JEunJin/BigData_python/refs/heads/master/bigdata_csvfile/gym_members_exercise_tracking.csv')
#data.info() # 결측치 없음 , 범주형변수 (Gender ,Workout_Type)

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Gender'] = le.fit_transform(data['Gender'])
data['Workout_Type'] = le.fit_transform(data['Workout_Type'])

train = data.iloc[:700,:]
test = data.iloc[700:,:]
#print(train.info())

from sklearn.model_selection import train_test_split
X = train.drop(columns='Experience_Level')
y = train['Experience_Level']
X_train , X_val , y_train, y_val = train_test_split(X,y,test_size=0.2)

from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=150 , max_depth=8)
rfc.fit(X_train,y_train)
y_pred = rfc.predict(X_val)

from sklearn.metrics import f1_score
result1 = f1_score(y_pred,y_val, average='macro')
print(result1) # 0.87 매우 정확함 
```

```python
X_test = test.drop(columns = 'Experience_Level')

y_pred2 = rfc.predict(X_test)

pd.DataFrame({'pred':y_pred2}).to_csv('9th_test_type2.csv',index=False)
print(pd.read_csv('9th_test_type2.csv'))
```

### 문제 (1)
- 주택가격_만원 예측
1. train 데이터를 활용하여 다중선형회귀분석 진행, 유의미한 독립변수의 개수를 구하시오 
2. 1번 문제에서 확인된 유의미한 독립변수들만을 사용해서 다중선형회귀분석 수행 , 해당 모델로 예측한 값의 피어슨 상관계수를 구하시오
3. 2번 문제에서 생성된 다중선형회귀분석 모델에 test 데이터를 적용 , 종속변수를 예측 , 예측값의 RMSE 수치를 구하시오 
(train 800행 , test 200행  , 모든 소수점은 3자리까지만 출력력)

```python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/JEunJin/BigData_python/refs/heads/master/bigdata_csvfile/real_estate_data.csv')
#data.info() # 결측치 없음 , 범주형변수 (Gender ,Workout_Type)
train = data.iloc[:800,:]
test =  data.iloc[800:,:]
train.head()

import statsmodels.api as sm
X = train.drop(columns='주택가격_만원')
y = train['주택가격_만원']

X1 = sm.add_constant(X)
result1 = sm.OLS(y,X1).fit().summary()
print(result1) 
# 문제 1 : 유의미한 독랍변수 개수 2개
print(2)
# 문제 2 유의미한 독립변수들로만 다중선형회귀분석 수행 , 예측한 값의 피어슨 상관계수
X2 = train[['주택면적_제곱미터','건축년도']]
X2 = sm.add_constant(X2)
model = sm.OLS(y,X2).fit()
#print(model.summary())
pred = model.predict(X2)
from scipy.stats import pearsonr
pear,p_val = pearsonr(y,pred)
print(round(pear,3))

# 문제 3: 2번에서 생성된 다중선형~모델에 test 데이터를 적용 , 종속변수를 예측 , 예측값의 RMSE 수치를 구하시오 
X3 = test[['주택면적_제곱미터','건축년도']]
X3 = sm.add_constant(X3)
pred2 = model.predict(X3)

from sklearn.metrics import root_mean_squared_error
val = test['주택가격_만원']
rmse = root_mean_squared_error(pred2,val)
print(round(rmse,3))

```

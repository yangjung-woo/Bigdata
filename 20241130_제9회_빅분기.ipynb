### 문제 (1)
1. '대출 A' 와 '대출 B'칼럼을 합하여 '총대출액' 컬럼을 만든다 
2. 성별간 총 대출액 차이의 절대값이 가장 큰 지역 코드를 구하시오 


```python
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/JEunJin/BigData_python/refs/heads/master/bigdata_csvfile/loan_data.csv')
data.head()
data['총대출액'] = data['대출A']+data['대출B']

#**
result = data.groupby(['지역코드','성별'])[['총대출액']].sum() 
# 대괄호 2개 : 데이터프레임 반환
# 대괄호 1개 : Series
# print(result)
# result2 = result.unstack()
# 성별 << 같은 지역코드 성별간 차이를 알기 위헤 멀티 인덱스 컬럼을 unstack()화 
result2 = pd.pivot_table(data=result, index='지역코드',columns='성별',values = '총대출액')
print(result2)

result2['지역간 차이'] = abs(result2[0]-result2[1])
display(result2.sort_values('지역간 차이',ascending=False))
print(1000009974)
```

### 문제 (2)
1. 각 연도별로 각 범죄의 검거건수를 발생건수로 나눈 검거율을 계산
2. 연도별로 검거율이 가장 높은 범죄형태의 검거건수의 총합을 계산 


```python
import pandas as pd 
data = pd.read_csv('https://raw.githubusercontent.com/JEunJin/BigData_python/refs/heads/master/bigdata_csvfile/crime_data_2014_2020.csv')

발생건수 = data.iloc[::2].set_index('년도').drop(columns='구분')
검거건수 = data.iloc[1::2].set_index('년도').drop(columns='구분')

검거율 = 검거건수 / 발생건수
display(검거율)

total = 0
# 연도별로 검거율이 가장 높은 범죄형태의 검거건수 총합 계산 
for year in 검거율.index:
    highest = 검거율.loc[year].idxmax()# 검거율이 가장 높은 범죄명을 highest 에 대입
    total += 검거건수.loc[year , highest]

print(total)
```

### 문제 (3)
1. 결측값이 존재하는 컬럼 2개 중 '연봉'은 해당 컬럼의 평균으로 대체 , '근속년수'는 '부서명'과 '근무평가등급'이 같은 직원의 '근속년수' 평균으로 대체 
2. 연봉을 근속년수로 나눈 컬럼과 , 연봉을 고객만족점수으로 나눈 컬럼을 파생변수로 생성 , 각 파생변수를 기준으로 데이터를 정렬 
3. 두개의 파생변수를 기준으로 내림차순 정렬한 가각의 데이터에서 세번째로 높은 데아터의 '근속년수'를 a , 두번째로 높은데아터의 '교육참가시간수'를 b 라고 할때 a+b를 구하시오ㅓ 
(단, 출력값을 정수 형태로 제출)

```python
import pandas as pd 
data = pd.read_csv('https://raw.githubusercontent.com/JEunJin/BigData_python/refs/heads/master/bigdata_csvfile/employee_data_with_grades.csv')
#data.head()
data['연봉'].fillna(data['연봉'].mean(),inplace=True)
# '부서명'과 '근무평가등급'에 대한 '근속년수'의 평균으로 결측값 채우기
data['근속년수'] = data.groupby(['부서명','근무평가등급'])[['근속년수']].transform(lambda x : x.fillna(x.mean())) # 각 그룹 내 평균값으로 결측값을 대체 

data.head()
data['연봉/근속년수'] = data['연봉']/data['근속년수']
data['연봉/고객만족점수'] = data['연봉']/data['고객만족점수']

data1 = data.sort_values('연봉/근속년수',ascending=False)
data2 = data.sort_values('연봉/고객만족점수',ascending=False)
display(data1.head()) # EMP0002 > 근속년수 : 1.0
display(data2.head()) # EMP0350 > 교육참가시간수 : 80
a = data1['근속년수'].iloc[2]
b = data2['교육참가시간수'].iloc[1]
print(int(a+b))
```
